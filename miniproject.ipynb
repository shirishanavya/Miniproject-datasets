### **Splitting data into X and Y**

```python
df['day_1']=df['units_sold'].shift(-1)
df['day_2']=df['units_sold'].shift(-2)
df['day_3']=df['units_sold'].shift(-3)
df['day-4']=df['units_sold'].shift(-4)
```
